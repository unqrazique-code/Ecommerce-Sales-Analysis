## Monthly Sales Analysis
This section shows the trend of sales over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/sales.csv")
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

Data Overview

```python
df.info()
df.describe()
```

## Monthly Sales Trend

```python
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

## Top Selling Products

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)

top_products.head(5)
```

## Sales by Region

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

## Profit by Category

```python
category_profit = df.groupby('Category')['Profit'].sum()

category_profit.plot(kind='pie', autopct='%1.1f%%')
plt.title("Profit Distribution")
plt.show()
```

## Key Insights
- Sales show a consistent trend over months  
- Technology products generate highest revenue  
- West region contributes maximum sales  
- Profit varies across categories
